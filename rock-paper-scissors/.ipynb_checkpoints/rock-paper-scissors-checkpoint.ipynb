# Rock Paper Scissors

Let's play the famous game against our computer.

https://en.wikipedia.org/wiki/Rock%E2%80%93paper%E2%80%93scissors

The use of functions is recommended

## Goals
1. Use of loop
2. Data capture by console
3. Use if-elif-else
4. Use of try-except
5. Definition of functions. Modular programming
6. Logical operators.
7. Print
8. Import modules

```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list

import random
userscore =0
computerscore = 0

choices = ["r","p","s"]
rnd = 1
while True:
    print ("Round", rnd)
    userchoice= input("What do you choose: Paper(p), Rock (r), Scissors (s) ?")
    compchoice = random.choice (choices)
    print("Computer chose:", compchoice)
    
    if len (userchoice) == 0:
        print ("Invalid answer, Try again!")
        continue

    elif userchoice [0].lower() == compchoice:
        print ("it's a Draw")
        continue
    
    elif userchoice [0].lower() == "r" and compchoice == "s":
        print ("User wins this round")
        userscore +=1
    elif userchoice [0].lower() == "r" and compchoice == "p":
        print ("Computer wins this round")
        computerscore +=1
    elif userchoice [0].lower() == "s" and compchoice == "p":
        print ("User wins this round")
        userscore +=1
    elif userchoice [0].lower() == "s" and compchoice == "r":
        print ("Computer wins this round")
        computerscore +=1
    elif userchoice [0].lower() == "p" and compchoice == "r":
        print ("Computer wins this round")
        computerscore +=1
    elif userchoice [0].lower() == "p" and compchoice == "s":
        print ("User wins this round")
        userscore +=1
    else:
        print ("Invalid Answer, Try again!")
        continue
    
    rnd = rnd +1
    if rnd == 5:
        break
print ("User Wins", userscore)
print ("Computer Wins", computerscore)

def wins():
    if userscore > computerscore:
        return ("User wins!")
    elif computerscore > userscore:
        return ("Computer wins!")
    else:
        return ("It's a Draw")
     
```
